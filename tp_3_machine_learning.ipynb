## import the data

```python
import pandas as pd
import numpy as np

from pandas import Series, DataFrame
from sklearn.grid_search import GridSearchCV
from sklearn import cross_validation, metrics

import matplotlib.pylab as plt
%matplotlib inline
```

```python
# read csv file

df_train = pd.read_csv("train.csv", sep=";")
df_train
```

```python
# df_train.info()
# df_train.describe()
```

```python
df_train['mariage'] = df_train['marital'].map({
    'married': 1, 'single': 2, 'divorced': 3, 'unknown': 0
}).astype(int)

df_train['work'] = df_train['job'].map({
    'blue-collar': 1, 'services': 2, 'admin.': 3, 'entrepreneur': 4,
    'self-employed': 5, 'technician': 6, 'management': 7, 'student': 8, 
    'retired': 9, 'housemaid': 10, 'unemployed': 11, 'unknown': 0
}).astype(int)

df_train['edu'] = df_train['education'].map({
    'basic.9y': 1, 'high.school': 2, 'university.degree': 3,
    'professional.course': 4, 'basic.6y': 5, 'basic.4y': 6, 
    'illiterate': 7,
    'unknown': 0
}).astype(int)

df_train['def'] = df_train['default'].map({
    'no': 1, 'yes': 2,
    'unknown': 0
}).astype(int)

df_train['house'] = df_train['housing'].map({
    'yes': 1, 'no': 2, 'unknown': 0
}).astype(int)

df_train['ploan'] = df_train['loan'].map({
    'yes': 1, 'no': 2, 'unknown': 0
}).astype(int)

df_train['outcome'] = df_train['poutcome'].map({
    'nonexistent': 2, 'failure': 0, 'success': 1
}).astype(int)

df_train['agreed'] = df_train['y'].map({
    'yes': 1, 'no': 0
}).astype(int)
```
